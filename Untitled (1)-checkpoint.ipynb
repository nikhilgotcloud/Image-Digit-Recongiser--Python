```python
import keras
from keras.datasets import mnist   #Importing Dataset
from keras.models import Sequential  #for sequential layer

from keras.layers import Dense, Dropout, Flatten  # for dense dropout and flatten layer
from keras.layers import Conv2D, MaxPooling2D # for convolutional and maxpooling Layer

from keras import backend as K

```

```python
#Splitting the Dataset #
```

```python
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print(x_train.shape, y_train.shape)  # prinbtung the shape of data 

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)  # reshaping into array of 4d so that it could be passes into model
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)

```

```python
# convert class vectors to binary class matrices
#Encoding Data

y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

```

```python
batch_size = 128
num_classes = 10

#declaring model and layers 

model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5),activation='relu',input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,optimizer=keras.optimizers.Adadelta(),metrics=['accuracy'])

```

```python
epochs = 20
```

```python
hist = model.fit(x_train, y_train,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(x_test, y_test))



```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

```

```python
model.save('model.h5')
print("model saved as model.h5")
```
